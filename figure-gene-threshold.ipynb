# Example of global thresholds for gene expression

Context-specific reconstruction methods usually rely on a prior classification of genes into higly and lowly expressed genes. Simplest method for that is based on global thresholds, i.e., quantile-based thresholds on the distribution of the gene expression (across all genes)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
sns.set(font_scale=1.25)
```

Example using data from "Low oxygen levels as a trigger for enhancement of respiratory metabolism in Saccharomyces cerevisiae", https://link.springer.com/article/10.1186/1471-2164-10-461. GEO GSE12442

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10, 6))
df_expr = pd.read_csv('./data/yeast/rintalaYeast2009_transcriptomics.csv')

# Condition of 20.9% of oxygen
sns.distplot(df_expr.O2_209, bins=60, ax=ax);
ax.set_xlabel('Normalized gene expression value')
ax.axvline(np.quantile(df_expr.O2_209, 0.10), linestyle='--')
ax.axvline(np.quantile(df_expr.O2_209, 0.90), linestyle='--')
fig.savefig('figures/thresholds.pdf', format='pdf')
```
