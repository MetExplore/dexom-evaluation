```python
%matplotlib inline

import matplotlib.lines as mlines
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set_style("whitegrid")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
pd.DataFrame({'Cholesterol': np.random.lognormal(0.5, 0.1, 100),
              'Fatty-acids': np.random.lognormal(0.7, 0.05, 100),
              'BCAAs metabolism': np.random.lognormal(0.25, 0.15, 100),
              'Glycolysis/Gluconeogenesis': np.random.lognormal(0.05, 0.05, 100)}).plot.box(ax=ax, 
                                                                                            boxprops=dict(linewidth=2.5),
                                                                                            medianprops=dict(linewidth=2.5));
ax.axhline(-np.log10(0.05), linestyle='--', color='k')
ax.set_xlabel('Pathways', fontsize=14)
ax.set_ylabel('-log10(p-value) (pathway enrichment)', fontsize=14);
```
